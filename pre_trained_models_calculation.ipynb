```python
import pandas as pd
import numpy as np
## !pip install sentence-transformers
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model_names = [
    "sentence-transformers/all-MiniLM-L6-v2",
    "sentence-transformers/all-mpnet-base-v2",
    "allganize/msmarco-distilbert-cos-v5_en-ko-ja_v1.2",
    "DangFutures/fine_bge_fed",
    "zoukagh/bethiz-bert-similarity-sentences2",
    "LazarusNLP/all-indobert-base-v2"
]
```

```python
parameters = ["cosine_similarity", "euclidean_distance", "manhattan_distance", "minkowski_distance", "correlation_coefficient"]

data = []

paragraph1 = """
The scientific community, driven by an insatiable curiosity to unravel the mysteries of the natural world,
conducts experiments, collects data, and formulates hypotheses with the overarching goal of expanding our understanding
of the intricate processes governing the cosmos, from the microscopic realms of quantum physics to
the vast expanses of the observable universe.
"""

paragraph2 = """

Motivated by an unquenchable curiosity to decipher the enigmas of the natural world, the scientific
community engages in a continuous process of experimentation, data collection, and hypothesis formulation.
Their ultimate aim is to enhance our comprehension of the intricate mechanisms that govern the cosmos,
spanning from the minute intricacies of quantum physics to the expansive reaches of the observable universe.
"""
```

```python
for model_name in model_names:
    model = SentenceTransformer(model_name)

    cosine_sim = cosine_similarity(model.encode([paragraph1]), model.encode([paragraph2]))[0][0]
    euclidean_dist = np.linalg.norm(model.encode([paragraph1]) - model.encode([paragraph2]))
    manhattan_dist = np.abs(model.encode([paragraph1]) - model.encode([paragraph2])).sum()
    minkowski_dist = np.power(np.power(np.abs(model.encode([paragraph1]) - model.encode([paragraph2])), 3).sum(), 1/3)
    jaccard_sim = len(set(paragraph1.split()) & set(paragraph2.split())) / len(set(paragraph1.split()) | set(paragraph2.split()))
    correlation_coeff = np.corrcoef(model.encode([paragraph1])[0], model.encode([paragraph2])[0])[0, 1]

    parameter_values = [cosine_sim, euclidean_dist, manhattan_dist, minkowski_dist, correlation_coeff]

    data.append([model_name] + parameter_values)
```

```python
columns = ["Model"] + parameters
df = pd.DataFrame(data, columns=columns)
df
```

```python
weights = np.array([1,1,1,1,1])
impacts= np.array([1,1,1,1,1])   ## all impacts are positive
```

```python
def topsis(df, parameters, criteria_weights):
    df_normalized = df.copy()

    # Normalize the DataFrame based on specified parameters
    for param in parameters:
        df_normalized[param] = (df[param] - df[param].min()) / (df[param].max() - df[param].min())

    # Multiply normalized matrix by criteria weights
    weighted_normalized_matrix = df_normalized.iloc[:, 1:] * criteria_weights

    # Calculate positive and negative ideal solutions
    positive_ideal_solution = weighted_normalized_matrix.max(axis=0)
    negative_ideal_solution = weighted_normalized_matrix.min(axis=0)

    # Calculate distances to positive and negative ideal solutions
    distance_positive_ideal = np.linalg.norm(weighted_normalized_matrix - positive_ideal_solution, axis=1)
    distance_negative_ideal = np.linalg.norm(weighted_normalized_matrix - negative_ideal_solution, axis=1)

    # Calculate TOPSIS score for each tuple
    df_normalized["TOPSIS_Score"] = np.sqrt(np.sum((weighted_normalized_matrix - positive_ideal_solution) ** 2, axis=1))

    # Rank the tuples based on TOPSIS score
    df_ranked = df_normalized.sort_values(by="TOPSIS_Score", ascending=False).reset_index(drop=True)

    return df_ranked
```

```python
df_result = topsis(df, parameters, weights)
df_result
```

```python
df_result.to_csv("topsis_results.csv", index=False)
```
